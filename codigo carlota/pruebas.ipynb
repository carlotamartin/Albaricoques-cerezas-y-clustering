Importamos los módulos que vamos a utilizar

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import  random as rdm
```

Tenemos que generar un dataset de frutas, para ello vamos a crear nuestros datos con números aleatorios mientras estos datos estén normalizados

```python
caracteristicasCerezas = [[17,19,1,5],[20,21,5,6],[22,23,6,7],[24,25,7,8.5],[26,27,8.5,10],[28,29,10,11.5]]

caracteristicasAlbaricoques = [[40,44,41],[45,49,54],[50,54,74],[55,59,100]]

```

Estamblecemos la cantidad de filas que va a haber en el dataset

```python
# [DIAMETRO, PESO]
cantidadObservaciones = 5000
```

Vamos a generar 5000 albaricoques y 500 cerezas

Creamos las cerezas

```python
cerezas = []
rdm.seed()
for iteration in range(cantidadObservaciones):
    #elección al azar de una característica
    cereza = rdm.choice(caracteristicasCerezas)
    #Generación de un diámetro
    diametro = round(rdm.uniform(cereza[0], cereza[1]),2)
    #Generación de un peso
    peso = round(rdm.uniform(cereza[2], cereza[3]),2)
    print ("Cereza "+str(iteration)+" "+str(cereza)+" : "+str(diametro)+" - "+str(peso))
    cerezas.append([diametro,peso])
```

Como vemos nos ha formado 5000 datos aleatorios en función de la lista que hemos entregado antes. Todos estos datos NORMALIZADOS

Ahora hacemos lo mismo para los albarizoques

```python
albaricoques = []
rdm.seed()
for iteration in range(cantidadObservaciones):
    #elección al azar de una característica
    albaricoque = rdm.choice(caracteristicasAlbaricoques)
    #Generación de un diámetro
    diametro = round(rdm.uniform(albaricoque[0], albaricoque[1]),2)
    #Generación de un peso
    limiteMinPeso = albaricoque[2] / 1.10
    limiteMaxPeso = albaricoque[2] * 1.10
    peso = round(rdm.uniform(limiteMinPeso, limiteMaxPeso),2)
    print ("Albaricoque "+str(iteration)+" "+str(albaricoque)+" : "+str(diametro)+" - "+str(peso))
    albaricoques.append([diametro,peso])
```

Creamos las observaciones

```python
frutas = cerezas+albaricoques
print(frutas[1:40])
```

Mezclamos las observaciones

```python
dataFrame = pd.DataFrame(frutas)
dataFrame.to_csv("datas/frutas.csv", index=False,header=False)
```

Ahora en estos datos vamos a emplear un algortimo del aprendizaje no supervisado. Vamos a hacer un k.means. Nos permite hacer clasificaciones en un acantidad de grupos de predicción conocida y en una cantidad pequeña de datos (<1000).

La fragmentación de datos, tamben llambada clustering, es distribuir los distintos datos en algo que tienen algo en común. Nosotros nos encargaremos de dar nombre a esos grupos y la maquina será la que los divida

### Visualización de datos

Vamos a cargar los archivos y a visualizar los datos de una forma genérica

```python
#Carga de los datos
frutas = pd.read_csv("datas/frutas.csv",
names=['DIAMETRO','PESO'], header=None)
```

```python
frutas.head(5)
```

```python
#Visualizar los datos
frutas.plot.scatter(x="DIAMETRO",y="PESO")
plt.show()
```

### Clasificamos los datos

```python
#Importamos las librerías

from sklearn.cluster import KMeans
```

```python
modelos = KMeans(n_clusters=2) #número de clusters
```

```python
#entrenamos las pruebas
modelos.fit(frutas)
```

```python
#Predicciones
predicciones_kmeans = modelos.predict(frutas)
```

```python
#Visualización de la clusterización
plt.scatter(frutas.DIAMETRO, frutas.PESO, c=predicciones_kmeans, s=50, cmap='viridis')
plt.xlabel("DIAMETRO")
plt.ylabel("PESO")

#Visualización de los centroides
centers = modelos.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.show()
```
